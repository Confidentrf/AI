```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow  # Use this for image display in Colab

# Load the image
image_color = cv2.imread('truckrr.png')

# Check if the image was loaded successfully
if image_color is None:
    print("Error: Image not found or failed to load.")
else:
    # Display the image using cv2_imshow (for Colab)
    cv2_imshow(image_color)

# No need for cv2.waitKey() or cv2.destroyAllWindows() in Google Colab
```

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

```python
image_gray.shape
```

```python
# Set thresholds for Canny edge detection
threshold_1 = 20
threshold_2 = 250

# Perform Canny edge detection
image_canny = cv2.Canny(image_gray, threshold_1, threshold_2)

# Display the Canny edge-detected image using cv2_imshow (for Colab)
cv2_imshow(image_canny)

# No need for cv2.waitKey() or cv2.destroyAllWindows() in Colab
```

```python
# Create FAST Detector object
fast = cv2.FastFeatureDetector_create()

# Detect keypoints in the grayscale image
keypoints = fast.detect(image_gray, None)

# Print the number of keypoints detected
print("Number of keypoints Detected: ", len(keypoints))

# Draw keypoints on the color image
image_with_keypoints = cv2.drawKeypoints(image_color, keypoints, None, color=(255, 0, 0))

# Display the image with keypoints using cv2_imshow (for Colab)
cv2_imshow(image_with_keypoints)

# No need for cv2.waitKey() or cv2.destroyAllWindows() in Colab
```

```python
# Create ORB object, we can specify the number of key points we desire
orb = cv2.ORB_create()

# Detect keypoints in the grayscale image
keypoints = orb.detect(image_gray, None)

# Obtain the descriptors
keypoints, descriptors = orb.compute(image_gray, keypoints)

# Print the number of keypoints detected
print("Number of keypoints Detected: ", len(keypoints))

# Draw rich keypoints on the input image
image_with_keypoints = cv2.drawKeypoints(image_color, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Display the image with keypoints using cv2_imshow (for Colab)
cv2_imshow(image_with_keypoints)

# No need for cv2.waitKey() or cv2.destroyAllWindows() in Colab
```
